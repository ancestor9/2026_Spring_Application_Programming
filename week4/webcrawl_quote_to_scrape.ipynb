## **Web Crwaling and Scrapping 실습**

### Text mining


```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
import pandas as pd
pd.DataFrame(X.toarray(),
             columns = vectorizer.get_feature_names_out())
```

```python
vectorizer2 = CountVectorizer(analyzer='word', ngram_range=(2, 2))
X2 = vectorizer2.fit_transform(corpus)
vectorizer2.get_feature_names_out()

pd.DataFrame(X2.toarray(),
             columns = vectorizer2.get_feature_names_out())
```

## http://quotes.toscrape.com/ 에서 문장을 추출하여
### 1. 가장 많이 나오는 단어와 출현 빈도 구하기
### 2. 각 문장을 출현단어 여부에 따라 수치형 데이터로 변환하기(Countervector)

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://quotes.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

for quote in soup.find_all("span", class_="text"):
    print(quote.text)
```

```python
quotes = [quote.text for quote in soup.find_all("span", class_="text")]
quotes
```

```python
texts = []
for quote in soup.find_all("span", class_="text"):
    # 따옴표를 제거한 텍스트
    cleaned_text = quote.text.strip('“”')
    print(cleaned_text)
    texts.append(cleaned_text)
print(texts)
```

```python
# text mining을 위해서는 corpus를 이렇게 반드시 저장
texts = [ quote.text.strip('“”') for quote in soup.find_all("span", class_="text")]
texts
```

```python
from collections import Counter
import string

# 텍스트 전처리 및 단어 빈도 계산
def word_frequency(texts):
    # 모든 텍스트를 소문자로 변환하고 구두점 제거
    cleaned_texts = ' '.join(texts).lower().translate(str.maketrans('', '', string.punctuation))
    # 단어 빈도 계산
    word_counts = Counter(cleaned_texts.split())
    return word_counts

# 단어 빈도 계산
word_counts = word_frequency(texts)
print(word_counts)
```

## 10페이지 까지 인용문 읽기(gemini)

```python
import requests
from bs4 import BeautifulSoup
import time

base_url = "http://quotes.toscrape.com"
current_url = base_url
page_count = 1

# 1. 데이터를 담을 빈 리스트 생성
all_quotes = []

print("🚀 데이터 수집 및 리스트 저장 시작...")

while page_count <= 10:
    response = requests.get(current_url)
    soup = BeautifulSoup(response.text, "html.parser")

    # 현재 페이지의 모든 인용문 요소 찾기
    quotes = soup.find_all("span", class_="text")

    for quote in quotes:
        # 2. 리스트에 인용문 텍스트 추가
        all_quotes.append(quote.text)

    # 다음 페이지 이동 로직
    next_button = soup.find("li", class_="next")
    if next_button and page_count < 10:
        current_url = base_url + next_button.find("a")["href"]
        page_count += 1
        time.sleep(0.1) # 속도를 위해 대기 시간을 줄였습니다
    else:
        break

# 3. 결과 확인
print(f"\n✅ 수집 완료! 총 {len(all_quotes)}개의 항목이 리스트에 저장되었습니다.")
print("-" * 30)
print("처음 3개 미리보기:", all_quotes[:3])
```

```python
word_frequency(all_quotes)
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(all_quotes)
```

```python
pd.DataFrame(X.toarray(),
             columns = vectorizer.get_feature_names_out())
```
